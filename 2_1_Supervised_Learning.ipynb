```python
# Math library
import numpy as np
# Plotting library
import matplotlib.pyplot as plt
```

```python
# Create some input / output data
x = np.array([0.03, 0.19, 0.34, 0.46, 0.78, 0.81, 1.08, 1.18, 1.39, 1.60, 1.65, 1.90])
y = np.array([0.67, 0.85, 1.05, 1.0, 1.40, 1.5, 1.3, 1.54, 1.55, 1.68, 1.73, 1.6 ])

print(x)
print(y)
```

```python
# Define 1D linear regression model
def f(x, phi0, phi1):
  # TODO :  Replace this line with the linear regression model (eq 2.4)
  y = phi0 + phi1 * x

  return y
```

```python
# Function to help plot the data
def plot(x, y, phi0, phi1):
    fig,ax = plt.subplots()
    ax.scatter(x,y)
    plt.xlim([0,2.0])
    plt.ylim([0,2.0])
    ax.set_xlabel('Input, $x$')
    ax.set_ylabel('Output, $y$')
    # Draw line
    x_line = np.arange(0,2,0.01)
    y_line = f(x_line, phi0, phi1)
    plt.plot(x_line, y_line,'b-',lw=2)

    plt.show()
```

```python
# Set the intercept and slope as in figure 2.2b
phi0 = 0.4 ; phi1 = 0.2
# Plot the data and the model
plot(x,y,phi0,phi1)
```

```python
# Function to calculate the loss
def compute_loss(x,y,phi0,phi1):

  # TODO Replace this line with the loss calculation (equation 2.5)
 # Calculate the predictions
    y_pred = phi0 + phi1 * x  # Model predictions for each input x

    # Calculate the squared differences
    squared_errors = (y - y_pred) ** 2

    # Compute the loss as the mean of squared errors divided by 2
    loss = 0.5 * np.sum(squared_errors)
    return loss
```

```python
# Compute the loss for our current model
loss = compute_loss(x,y,phi0,phi1)
print(f'Your Loss = {loss:3.2f}, Ground truth =7.07')
```

```python
# Function to calculate the loss
def compute_loss(x, y, phi0, phi1):
    y_pred = f(x, phi0, phi1)  # Predicted y values
    squared_errors = (y - y_pred) ** 2  # Squared error for each point
    loss = 0.5 * np.sum(squared_errors)  # Total loss
    return loss

# Start with initial values for phi0 and phi1
phi0 = 1.60
phi1 = 0.5

# Plot the initial model
plot(x, y, phi0, phi1)
print(f'Initial Loss = {compute_loss(x, y, phi0, phi1):3.2f}')

# First, fix phi1 and adjust phi0 to minimize loss
phi0_adjusted = phi0
loss = compute_loss(x, y, phi0_adjusted, phi1)

# Manually adjust phi0 and plot the loss (you can try different values of phi0)
phi0_values = np.arange(1.0, 2.0, 0.05)  # Try different values of phi0
losses_phi0 = []

for phi0_val in phi0_values:
    current_loss = compute_loss(x, y, phi0_val, phi1)
    losses_phi0.append(current_loss)
    print(f"phi0 = {phi0_val:.2f}, Loss = {current_loss:.2f}")

# Plot the loss curve as phi0 changes
plt.plot(phi0_values, losses_phi0)
plt.xlabel('phi0 values')
plt.ylabel('Loss')
plt.title('Loss vs phi0 values')
plt.show()

# Once the best phi0 is found, fix phi0 and adjust phi1
phi0_final = phi0_values[np.argmin(losses_phi0)]  # Find the optimal phi0

# Now, adjust phi1 while keeping phi0 fixed
phi1_values = np.arange(-1.0, 1.0, 0.05)  # Try different values of phi1
losses_phi1 = []

for phi1_val in phi1_values:
    current_loss = compute_loss(x, y, phi0_final, phi1_val)
    losses_phi1.append(current_loss)
    print(f"phi1 = {phi1_val:.2f}, Loss = {current_loss:.2f}")

# Plot the loss curve as phi1 changes
plt.plot(phi1_values, losses_phi1)
plt.xlabel('phi1 values')
plt.ylabel('Loss')
plt.title('Loss vs phi1 values')
plt.show()

# Find the optimal phi1
phi1_final = phi1_values[np.argmin(losses_phi1)]  # Find the optimal phi1

# Print the final best parameters and plot the best-fitting line
print(f"Optimal phi0 = {phi0_final:.2f}, Optimal phi1 = {phi1_final:.2f}")
plot(x, y, phi0_final, phi1_final)
```

# Visualizing the loss function

The above process is equivalent to descending coordinate wise on the loss function<br>

Now let's plot that function

```python
# Make a 2D grid of possible phi0 and phi1 values
phi0_mesh, phi1_mesh = np.meshgrid(np.arange(0.0,2.0,0.02), np.arange(-1.0,1.0,0.02))

# Make a 2D array for the losses
all_losses = np.zeros_like(phi1_mesh)
# Run through each 2D combination of phi0, phi1 and compute loss
for indices,temp in np.ndenumerate(phi1_mesh):
    all_losses[indices] = compute_loss(x,y, phi0_mesh[indices], phi1_mesh[indices])
```

```python
# Plot the loss function as a heatmap
fig = plt.figure()
ax = plt.axes()
fig.set_size_inches(7,7)
levels = 256
ax.contourf(phi0_mesh, phi1_mesh, all_losses ,levels)
levels = 40
ax.contour(phi0_mesh, phi1_mesh, all_losses ,levels, colors=['#80808080'])
ax.set_ylim([1,-1])
ax.set_xlabel(r'Intercept, $\phi_0$')
ax.set_ylabel(r'Slope, $\phi_1$')

# Plot the position of your best fitting line on the loss function
# It should be close to the minimum
ax.plot(phi0,phi1,'ro')
plt.show()
```
